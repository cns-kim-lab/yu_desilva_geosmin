```python
import pandas as pd 
import sys
sys.path.append('../../code')
import plot_of_experiment_data 
import matplotlib.pyplot as plt
import matplotlib
import warnings
import numpy as np 
warnings.filterwarnings("ignore", message="findfont: Font family 'Arial' not found.")
matplotlib.rcParams['svg.fonttype'] = 'none'
matplotlib.rcParams['font.family'] = 'Arial'
```

```python
def std_err(arr):
    err = np.nanstd(arr,axis=0)/np.sqrt(np.sum(~np.isnan(arr),axis=0))
    return err

def pval_to_asterisks(p):
    if p < 0.0001:
        return "****"
    elif p < 0.001:
        return "***"
    elif p < 0.01:
        return "**"
    elif p < 0.05:
        return "*"
    else:
        return "ns"
```

```python
### figure C: Optogenetics Or56a
fpath = '/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/inhibitory_neuron_optogenetics/optogenetics_geosmin_pathway.xlsx'
df = pd.read_excel(fpath,sheet_name='Or56a')
data,pivot,error = plot_of_experiment_data.make_pivot_optogenetics_data(df)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(1.8, 1.5))

ax,stat_all = plot_of_experiment_data.draw_paired_bar_figure(
    data=data,
    mean_df=pivot,
    err_df=error,
    ax=ax,
    conds=[0, 1],
    groups=['retinal (-)', 'retinal (+)'],
    get_values=lambda data, cond, group: data[(cond, 0 if group == 'retinal (-)' else 1)],
    stat_pairs=[
        ((0, 'retinal (+)'), (1, 'retinal (+)'))
    ],
    colors=('#AEACAB', 'black'),
    ylabel='Tarsal PER',
    xtick_func=lambda c: '-' if c == 0 else '+',
    ylim=(0, 0.89),
    yticks=np.round(np.arange(0, 1.25, .25), 2)
)
```

```python
import pandas as pd 
import sys
sys.path.append('../code')
import plot_of_experiment_data 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", message="findfont: Font family 'Arial' not found.")
```

```python
##### data load #####
## dose dependent PER 
# fig1a. 
tarsal_s_df = pd.read_excel('/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/sugar+geomsin/tarsal_PER.xlsx',index_col=False,sheet_name='sugar')
tarsal_sg_df = pd.read_excel('/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/sugar+geomsin/tarsal_PER.xlsx',index_col=False,sheet_name='sugar+geosmin')

# fig1b
labial_s_df = pd.read_excel('/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/sugar+geomsin/labial_PER.xlsx',index_col=False,sheet_name='sugar')
labial_sg_df = pd.read_excel('/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/sugar+geomsin/labial_PER.xlsx',index_col=False,sheet_name='sugar+geosmin')


## receptor screening 
# fig1d-e 
fpath = '/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/tarsal_receptor/tarsal_sugar_receptor.xlsx'

sheets = pd.ExcelFile(fpath)
sheets = sheets.sheet_names
# sheets = ['atGRN+','atGRN>Kir2.1']
data_df_receptor = {}
for s in sheets:
    df = pd.read_excel(fpath,s)
    data_df_receptor[s] = df

fpath = '/volume_4/research/seongbong/flywire/geosmin_project/LM_data_from_dilan/experiment/tarsal_receptor/tarsal_sugar_receptor_with_geosmin.xlsx'

sheets = pd.ExcelFile(fpath)
sheets = sheets.sheet_names
data_df_receptor_geosmin = {}
for s in sheets:
    df = pd.read_excel(fpath,s)
    data_df_receptor_geosmin[s] = df
    
```

# fig1a-b: sucrose dose dependent PER 

```python
# fig1a: sucrose on leg 
fig,ax = plt.subplots(1,1,figsize=(1.8,2))
plot_of_experiment_data.plot_dose_dp_PER(ax,tarsal_s_df)
plot_of_experiment_data.plot_dose_dp_PER(ax,tarsal_sg_df,cidx=1)
stat_all = plot_of_experiment_data.plot_statistics_result(ax,[tarsal_s_df,tarsal_sg_df])

ax.set_ylabel('Tarsal PER',fontsize=10)
ax.set_xlabel('[sucrose] (mM)',fontsize=10)
ax.set_ylim([0,0.8])
ax.set_xticks(list(range(len(tarsal_s_df.columns))),[x[:-2] for x in tarsal_s_df.columns],fontsize=10)
```

```python
stat_all
```

```python
# fig1b: sucrose on labellum  
fig,ax = plt.subplots(1,1,figsize=(1.8,2))
plot_of_experiment_data.plot_dose_dp_PER(ax,labial_s_df)
plot_of_experiment_data.plot_dose_dp_PER(ax,labial_sg_df,cidx=1)
stat_all = plot_of_experiment_data.plot_statistics_result(ax,[labial_s_df,labial_sg_df])

ax.set_ylabel('Labellar PER',fontsize=10)
ax.set_xlabel('[sucrose] (mM)',fontsize=10)
ax.set_ylim([0,0.8])
ax.set_xticks(list(range(len(labial_s_df.columns))),[x[:-2] for x in labial_s_df.columns],fontsize=10)
```

```python
stat_all
```

# fig1d-e: sucrose dose dependent PER in sugar receptor silenced flies 

```python
# fig1c
fig,ax = plt.subplots(1,1,figsize=(1.8,2))
conditions = ['control','Gr5a>Kir2.1','Gr64f>Kir2.1']
alpha = [.5,1,.5]
linestyle = ['-.','-',':']
shape =  ['o','v','s']
for i,condition in enumerate(conditions):
    plot_of_experiment_data.plot_dose_dp_PER(ax,data_df_receptor[condition],alpha=alpha[i],linestyle=linestyle[i],shape=shape[i])
    plot_of_experiment_data.plot_dose_dp_PER(ax,data_df_receptor_geosmin[condition],cidx=1,alpha=alpha[i],linestyle=linestyle[i],shape=shape[i])
    
ax.set_ylabel('Tarsal PER',fontsize=10)
ax.set_xlabel('[sucrose] (mM)',fontsize=10)
ax.set_ylim([0,0.8])
ax.set_xticks(list(range(len(data_df_receptor['control'].columns))),[x[:-2] for x in data_df_receptor['control'].columns],fontsize=10)
```

```python
# fig1d
fig,ax = plt.subplots(1,1,figsize=(1.8,2))
conditions = ['control','atGRN>Kir2.1','Gr64f>Kir2.1']
alpha = [.5,1,.5]
linestyle = ['-.','-',':']
shape =  ['o','v','s']
for i,condition in enumerate(conditions):
    plot_of_experiment_data.plot_dose_dp_PER(ax,data_df_receptor[condition],alpha=alpha[i],linestyle=linestyle[i],shape=shape[i])
    plot_of_experiment_data.plot_dose_dp_PER(ax,data_df_receptor_geosmin[condition],cidx=1,alpha=alpha[i],linestyle=linestyle[i],shape=shape[i])
    
ax.set_ylabel('Tarsal PER',fontsize=10)
ax.set_xlabel('[sucrose] (mM)',fontsize=10)
ax.set_ylim([0,0.8])
ax.set_xticks(list(range(len(data_df_receptor['control'].columns))),[x[:-2] for x in data_df_receptor['control'].columns],fontsize=10)
```
